```python
def flatten_dict(d, parent_key='', sep='.'):
    items = []
    if isinstance(d, list):
        for i, v in enumerate(d):
            if isinstance(v, dict):
                items.extend(flatten_dict(v, f"{parent_key}{sep}{i}", sep).items())
            else:
                items.append((f"{parent_key}{sep}{i}", v))
    elif isinstance(d, dict):
        for k, v in d.items():
            new_key = f"{parent_key}{sep}{k}" if parent_key else k
            if isinstance(v, dict):
                items.extend(flatten_dict(v, new_key, sep).items())
            elif isinstance(v, list):
                items.extend(flatten_dict(v, new_key, sep).items())
            else:
                items.append((new_key, v))
    return dict(items)
```

```python
def sort_dict(d):
    return {key: d[key] for key in sorted(d)}
```

```python
import json

# Load the JSON data from a file
with open('pokedex.json') as f:
    data = json.load(f)

# Flatten the dictionary
flattened_data_arr = [flatten_dict(data[i]) for i, v in enumerate(data)]

# flattened_data_arr = [sort_dict(flatten_dict(data[i])) for i, v in enumerate(data)]
flattened_data_arr
# Print the flattened dictionary
# print(flattened_data)
```

```python
from collections import defaultdict
stacked = defaultdict(list)
for d in flattened_data_arr:
    for k, v in d.items():
        stacked[k].append(v)

print(dict(stacked))
```

```python
height_in_meters = []

for h in stacked['height']:
    start_index = h.find("(")
    end_index = h.find("m")
    if start_index != -1 and end_index != -1:  # height has parentheses
        height_in_meters.append(h[start_index+1:end_index])
print(height_in_meters)
stacked['height'] = height_in_meters
```

```python
weight_in_meters = []

for h in stacked['weight']:
    start_index = h.find("(")
    end_index = h.find(" kg")
    if start_index != -1 and end_index != -1:
        weight_in_meters.append(h[start_index+1:end_index])
print(weight_in_meters)
stacked['weight'] = weight_in_meters
```
