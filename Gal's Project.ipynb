```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from statsmodels.graphics.gofplots import qqplot
from scipy.stats import norm
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression
from IPython import display
```

```python
df = pd.read_csv(r'gal_scoring.csv')
df
```

```python
plt.plot(df.age,df.score,marker='o',markerfacecolor='blue')
plt.xticks(df.age,rotation='vertical')
plt.xlabel('Age')
plt.ylabel('Score')
```

```python
from scipy.stats import spearmanr

r, pvalue = spearmanr(df['age'], df['score'])
sns.regplot(data=df,x='age',y='score',label=f'spearmanr = {r:.2f}')
plt.xticks(df.age,rotation='vertical')
plt.legend()
```
